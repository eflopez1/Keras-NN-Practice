```python
import numpy as np
import pandas as pd

#dataset import
dataset = pd.read_csv('train.csv')
dataset.head(10) #Returns first 10 rows of data
```

```python
#Change pandas dataframe to numpy array
X = dataset.iloc[:,:20].values
y = dataset.iloc[:,20:21].values
```

```python
#Normalize the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
#Must OneHot encode the classes so it appears as:
#0 = 1000
#2 = 0100
#3 = 0010
#4 = 0001
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
y = ohe.fit_transform(y).toarray()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.1, random_state=0)

#That is it for data prep! Now let's make a Neural Network
```

```python
print("Length X_Train:", len(X_train))
print("Length Y_train:", len(y_train))
```

```python
import tensorflow.keras as keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

#Building Neural Network
model = Sequential()
model.add(Dense(16,input_dim=20, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(4, activation='softmax'))

#Make the loss function and optimizer
model.compile(loss='categorical_crossentropy', optimizer='adam')
metrics=(['accuracy'])
```

```python
model.summary()
```

```python
# Train the model:
history = model.fit(X_train, y_train, epochs=100, batch_size=64)
```

```python
y_pred = model.predict(X_test)

#Converting predictions to label
pred = list()
for i in range(len(y_pred)):
    pred.append(np.argmax(y[i]))

# Converting on hot encoded test label to label
test = list()
for i in range(len(y_test)):
    test.append(np.argmax(y_test[i]))
```

```python
from sklearn.metrics import accuracy_score
a = accuracy_score(pred, test)
print('Accuracy is:', a*100)
```

```python
history = model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs=100, batch_size=64)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train','Test'], loc='upper left')
plt.show()
```
